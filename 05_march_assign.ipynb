```python
"""Q1. How can you create a Bokeh plot using Python code?"""

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show

x = [1,2,3,4,5,6,7,8,9]
y = [1,4,9,16,25,36,49,64,81]

output_file('test.html')
p = figure(title = 'Square of numbers')
p.xaxis.axis_label = 'Numbers'
p.yaxis.axis_label = 'Square of numbers'
p.circle(x, y)
show(p)
```

```python
"""Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example."""

"""Ans: Glyphs in Bokeh are the visual marks that are used to represent data in a plot. Glyphs can be simple geometric shapes, 
        such as circles, squares, or triangles, or more complex shapes, such as lines or patches.
        
        To add glyphs to a Bokeh plot, you first need to create a figure object using the figure() function. Then, you can add one 
        or more glyphs to the figure using a glyph method, such as line(), circle(), or square(), depending on the type of glyph you want to add."""


# Example: 
x = [1,2,3,4,5,6,7,8,9]
y = [1,4,9,16,25,36,49,64,81]

output_file("glyphs.html")
p = figure(title="Glyphs example", x_axis_label='x', y_axis_label='y')
p.line(x, y, legend_label="Line")
p.circle(x, y, legend_label="Circle")
p.square(x, y, legend_label="Square")
show(p)
```

```python
"""Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?"""

# Here are some of the most common ways to customize a Bokeh plot:
# 1. Title: You can add a title to your Bokeh plot using the title attribute of the figure object. For example:
p = figure(title="My Bokeh Plot")
# 2. Axes: You can customize the appearance of the x-axis and y-axis using the xaxis and yaxis attributes of the figure object. For example, to change the font size of the x-axis and y-axis labels, you can do:
p.xaxis.axis_label_text_font_size = "14pt"
p.yaxis.axis_label_text_font_size = "14pt"
# 3. Legend: You can customize the appearance of the legend using the legend attribute of the figure object. For example, to change the location of the legend, you can do:
p.legend.location = "top_left"
# 4. Colors: You can customize the color of various elements in your Bokeh plot, such as the line color and fill color of glyphs, using the line_color, fill_color, and color attributes of the glyph methods. For example:
p.line(x, y, line_color="red")
p.circle(x, y, fill_color="blue")
p.square(x, y, color="green")
```
